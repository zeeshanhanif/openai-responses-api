```python
import os
from openai import OpenAI
from dotenv import load_dotenv
import rich
```

```python
load_dotenv()

api_key = os.getenv('OPENAI_API_KEY')
MODEL = "gpt-4o-mini"

openai = OpenAI()
```

Search the web for the latest information before generating a response

1) In Chat API, model always retrieves information from web before responding to your query
2) In Responses API, you enable web search by configuring it in the tools, and model can choose to serach the web or not based on the content of the input prompt.

In the Responses API, if you want to ensure that the model must call the tool, you can use the `tool_choice` parameter and set it to `{type: "web_search_preview"}`. This helps reduce latency and provides more consistent results.

Since the Chat API uses a specific model that always performs a web search, this use case is more relevant to the Responses API.

# Responses API

https://platform.openai.com/docs/guides/tools-web-search?api-mode=responses

```python
response = openai.responses.create(
    model=MODEL,
    tools=[{
        "type": "web_search_preview",
        "user_location": {
            "type": "approximate",
            "country": "PK",
            "city": "Lahore",
            "region": "Punjab",
        }
    }],
    # input="List the best restaurants for Seafood",
    input="What is Agentic AI?",
    tool_choice={"type": "web_search_preview"} # Added tool_choice property
)

print(response.output_text)
rich.print(response.output)
rich.print(response.tools)
# rich.print(response)
```
